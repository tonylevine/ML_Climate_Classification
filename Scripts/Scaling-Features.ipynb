```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.preprocessing import StandardScaler 
import os
```

```python
df = pd.read_csv("~/Data Sets/Unsupervised/Dataset-weather-prediction-dataset-processed.csv")
```

```python
# Check DataFrame
# Very wide format - can be problematic
df.info()
```

```python
df.head()
```

```python
# Create scaler object
scaler = StandardScaler()
```

```python
# Apply StandardScaler to all columns
df_scaled = pd.DataFrame(scaler.fit_transform(df))
df_scaled.head()
```

```python
# Check that mean ~0 and std ~1 for all variables
# Looks ok, but too many columns to properly confirm
df_scaled.describe()
```

```python
# Convert from wide format to long format using melt
df_long = df.melt(id_vars=['DATE', 'MONTH'], 
                 var_name='city_metric', 
                 value_name='value')

# Split city name from variables and pivot them into new columns
df_long[['city', 'metric']] = df_long['city_metric'].str.split('_', n=1, expand=True)
df_long = df_long.drop('city_metric', axis=1).pivot_table(
   index=['DATE', 'MONTH', 'city'], 
   columns='metric', 
   values='value', 
   aggfunc='first'
).reset_index()
```

```python
# Check results
df_long.info()
```

```python
df_long.head(19)
```

```python
# Identify columns to exclude from scaling
categorical_cols = ['DATE', 'MONTH', 'city']
numeric_cols = df_long.select_dtypes(include=[np.number]).columns.difference(categorical_cols)

# Apply StandardScaler only to numeric columns
df_long[numeric_cols] = scaler.fit_transform(df_long[numeric_cols])
```

```python
# Confirm mean is ~0 and std is ~1 for all numeric variables
df_long[numeric_cols].describe()
```

```python
df_long['DATE'] = pd.to_datetime(df_long['DATE'], format='%Y%m%d', errors='coerce')
```

```python
df_long.to_pickle("~/Data Sets/Unsupervised/Dataset_scaled.pkl")
```
